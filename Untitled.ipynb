```python
latent, classes, _ = compute_latent(inn, data.get_dataset(c.dataset), compute_grad=False)
m = latent.mean(dim=0)
```

```python
m = torch.zeros(c.n_classes, inn.latent_dim)
s = torch.zeros(c.n_classes, inn.latent_dim)

for i in range(c.n_classes):
    m[i] = latent[classes == i].mean(dim=0)
    s[i] = latent[classes == i].std(dim=0)

print(m.repeat_interleave(c.n_classes, dim=0))
```

```python
y = torch.tensor(
    list(range(c.n_classes)), device=device
).repeat_interleave(c.n_classes)
t = torch.normal(
    mean=m.repeat_interleave(c.n_classes, dim=0).cuda(), std=s.repeat_interleave(c.n_classes, dim=0).cuda()
)
# t = torch.randn(c.n_classes * c.n_classes,inn.latent_dim, device=device)
x = inn.sample(t, inn.labels2condition(y))

plt.figure(figsize=(12,12))
plt.imshow(data.tensors2image(c.dataset, x, c.n_classes).transpose(1, 2, 0))
plt.yticks(range(16, 250, 34), ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surprise'], fontsize=14)
```
